# Доминирование

```python
import numpy as np
from itertools import product
```

Рассмотрим игру с нулевой суммой с матрицей

$$
	A=\begin{pmatrix}
	-2 & 2 & -1 & 0 & 1 \\
	2 & 3 & 1 & 2 & 2 \\
	3 & -3 & 2 & 4 & 3 \\
	-2 & 1 & -2 & -1 & 0
	\end{pmatrix}
$$


```python
A = np.array([[-2 , 2 , -1 , 0 , 1], [2 , 3 , 1 , 2 , 2], [3 , -3 , 2 , 4 , 3], [-2 , 1 , -2 , -1 , 0]])
A
```

## Строгое доминирование для первого игрока

```python
# размер матрицы
n, m = A.shape

ind1 = []

for i, j in product( iter(range(n)), iter(range(n)) ):
	if np.array( A[i,:]<A[j,:], dtype=int).sum() == m :
		if len(ind1)==0 :
			ind1.append(i)
		elif ind1[-1]!= i:
			ind1.append(i)

ind1
```

## Строгое доминирование для второго игрока

```python
# размер матрицы
n, m = A.shape

ind2 = []

for i, j in product( iter(range(m)), iter(range(m)) ):
	if np.array( A[:,i]>A[:,j], dtype=int).sum() == n:
		if len(ind2)==0 :
			ind2.append(i)
		elif ind2[-1]!= i:
			ind2.append(i)

ind2
```

##  Слабое доминирование для первого игрока

```python
# размер матрицы
n, m = A.shape

ind1 = []

for i, j in product( iter(range(n)), iter(range(n)) ):
	if (np.array( A[i,:]<=A[j,:], dtype=int).sum() == m) & (np.array( A[i,:]==A[j,:], dtype=int).sum()<m):
		if len(ind1)==0 :
			ind1.append(i)
		elif ind1[-1]!= i:
			ind1.append(i)

ind1
```

##  Слабое доминирование для второго игрока

```python
# размер матрицы
n, m = A.shape

ind2 = []

for i, j in product( iter(range(m)), iter(range(m)) ):
	if (np.array( A[:,i]>=A[:,j], dtype=int).sum() == n) & (np.array( A[:,i]==A[:,j], dtype=int).sum()<n):
		if len(ind2)==0 :
			ind2.append(i)
		elif ind2[-1]!= i:
			ind2.append(i)

ind2
```

```python
# удалим строки с индексами
A = np.delete(A, ind1, axis=0)
```

```python
# удалим столбцы с индексами
A = np.delete(A, ind2, axis=1)
```
